```python
from collections import Counter
import re
```

```python
transcript = "Hello, this is my voice to text project. I am happy to learn NLP and AI. The weather is amazing today!"

print(" Transcript:", transcript)
```

```python
tokens = re.findall(r"\b[a-zA-Z']+\b", transcript.lower())
```

```python
stopwords = {"i","me","my","we","you","he","she","it","they","is","am","are","was","were","the","a","an","to","of","and","in","on","for","with","this","that","today"}
words = [w for w in tokens if w not in stopwords]

```

```python
freq = Counter(words).most_common(10)
print("\n Top Keywords:", freq)
```

```python
positive_words = {"happy","love","amazing","great","good","excited","wonderful"}
negative_words = {"sad","bad","angry","hate","terrible","worse","problem"}

pos_count = sum(1 for w in words if w in positive_words)
neg_count = sum(1 for w in words if w in negative_words)

if pos_count > neg_count:
    sentiment = " Positive"
elif neg_count > pos_count:
    sentiment = " Negative"
else:
    sentiment = " Neutral"

print("\n Sentiment:", sentiment)
```

```python
!pip install wordcloud matplotlib

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a WordCloud from your words
wc = WordCloud(width=800, height=400, background_color="white").generate(" ".join(words))

plt.figure(figsize=(10,5))
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```
